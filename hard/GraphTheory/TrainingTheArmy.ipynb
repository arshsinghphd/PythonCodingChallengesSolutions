The problem concerns constrained optimization, with the goal to make each class have non-zero elements. 

+ Number of skills N is such that 1<=N<=200.
+ Number of individuals with each skill is such that 0<=C<=10, so there can be 2000 individuals
+ There can be upto 30 wizards 
+ If we had to do a naive combination to make a list of possible final state we will create an array of upto 60,000 possibilities. 
+ To optimize this matching, I will create three classes:
    + Cadet: a cadet has a name and a skill. It's skill is mutable.
    + Wizard: a wizard has a name, two lists of skills A and B. 
        + A wizard can accept a cadet with skills in list A. 
        + It can turn this skill into a skills from the list B. 
        + If one such transformation is performed, the list A and B are changed to reflect that the ability of the wizard to accept the skill and provide new skills are each reduced by 1.
    + Army: 
        + an Army is a collection of cadets and wizards. 
        + Cadets are enumerated in dictionary Cadets;
        + wizards in dictionary Wizards. 
        + There also is a list of count of cadets by skills, called Skills, which is indexed by (skill-1). e.g. Skill 1 is indexed at 0. The cadets can be surveyed using the function surveySkills that returns this list for present status of the army.
        + A dictionary of cadets keyed by skills, called CadetSkills is also created/reset inside surveySkills()
        
+ Matching:
   + Make a list of skills known to more than one cadets, a f(rom) list.
   + Make a list of skills that no cadets have C_i==0, a t(o) list. 
   + Use these lists to create a dictionary of list of wizards that can perform the transformations keyed by elements of f,t list.
   + sort these by number of wizards and start the transformations. That is we start transformations that can be only perfomed by one wizard, to the ones that can be performed by numerous wizards, counting the skills that go from 0 to non-zero. When we finish trying to make each transformation, we can return this count as the answer.
   

```python
import math
import os
import random
import re
import sys

class Wizard:
    def __init__(self,key,A,B):
        self.key=key
        self.A=A
        self.B=B
    def getKey(self):
        return self.key
    def setA(self,A):
        self.A=A
    def setB(self,B):
        self.B=B
    def getA(self):
        return self.A
    def getB(self):
        return self.B
    def copy(self):
        k=self.getKey()
        A=self.getA()
        B=self.getB()
        return Wizard(k,A,B)
    
class Cadet:
    def __init__(self,name,skill):
        self.name=name
        self.skill=skill
    def getName(self):
        return self.name
    def getSkill(self):
        return self.skill
    def setSkill(self,newskill):
        self.skill=newskill
    
class Army:
    def __init__(self,W,C,N,S):
        self.Wizards=W
        self.Cadets=C
        self.NSkills=N
        self.CadetSkills=None
        self.Skills=S
        if self.surveySkills()==S:
            print('Army ready for optimization')
    def getCadets(self):
        return self.Cadets
    def getWizards(self):
        return self.Wizards
    def assign_wc_ns(self,w,c,ns):
        A=w.getA()
        B=w.getB()
        s=c.getSkill()
        if (s in A) and (ns in B):
            A.remove(s)
            B.remove(ns)
            c.setSkill(ns)
            w.setA(A)
            w.setB(B)
    def getNSkills(self):
        return self.NSkills
    def getSkills(self):
        return self.Skills
    def setSkills(self,NS):
        self.Skills=NS
    def surveySkills(self):
        C=self.getCadets()
        NSkills=self.getNSkills()
        NewSkills=[0 for i in range(NSkills)]
        d={}
        for i in range(N):
            d[i+1]=[]
        for name in C:
            i=C[name].getSkill()
            NewSkills[i-1]+=1
            d[i].append(C[name])
        self.CadetSkills=d
        return NewSkills
    def getCadetSkills(self):
        return self.CadetSkills
```
